```python
import math
def factorize(n):
    """
    - Implement the basic brute force factorization algorithm shown as pseudo code
      below to find prime factors of n.
    
    Pseudocode:
        # n is a number, return the smallest factor of n
        for i from 2 to n-1:
            if i divides n:
                return i
        return FALSE"""

    
# from Theorem 2, if n is a composite integer, n has a prime divisor less than or equal to sqrt(n)
   #   result = 0

    
    a = int(math.sqrt(n))           
    if n % 2 == 0:            #if n even return 2
        return 2
    i = 3                  # counter
    while i<=a:              
        if n % i == 0:      #while loop to determin prime factors  
            return i  
        i += 2               
    
print(factorize(979))
print(299//13)
```

```python
import math
def factorizebetter(n):
# from Theorem 2, if n is a composite integer, n has a prime divisor less than or equal to sqrt(n)
#reference https://www.geeksforgeeks.org/ for understanding and providing factors list and running efficient algorithm for prime factors .


    i = 2                    
    factors = []                 #perform divisions of n by successive primes beginning with 2
    while i**2 <= n:          # from Theorem 2, if n is a composite integer, n has a prime divisor less than or equal to sqrt(n)
        if n % i:
            i += 1         #increment if remainder
        else:
            n = n // i       # continue dividing by successive primes ; append to list
            factors.append(i)
    if n > 1:
        factors.append(n)
    return factors
print(factorizebetter(7007))
```
